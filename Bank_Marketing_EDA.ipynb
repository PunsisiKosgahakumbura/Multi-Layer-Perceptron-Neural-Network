```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Visualize the distribution of the target variable
sns.countplot(x='deposit', data=data)
plt.title('Distribution of Deposit')
plt.show()


```

```python
# Visualize the distribution of numerical variables
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_columns].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Variables', y=1.02)
plt.show()
```

```python
# Visualize the correlation matrix of numerical variables
correlation_matrix = data[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize the distribution of categorical variables
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=data, hue='deposit')
    plt.title(f'Distribution of {column}')
    plt.show()
```
